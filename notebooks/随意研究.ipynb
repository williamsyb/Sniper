```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import akshare as ak
%matplotlib inline
```

```python
stock_zh_a_daily_hfq_df = ak.stock_zh_a_daily(symbol="sz000002", start_date='20201103', end_date='20201116', adjust="hfq")
stock_zh_a_daily_hfq_df
```

```python
def split_data(origin_data) -> dict:
    datas = []
    times = []
    for i in range(len(origin_data)):
        datas.append(origin_data[i][1:])
        times.append(origin_data[i][0:1][0])
    return {
        "datas": datas,
        "times": times,
    }
from sniper.plot.candlestick import Candle
df=stock_zh_a_daily_hfq_df.reset_index()
df.date=df.date.apply(lambda x:x.strftime('%Y-%m-%d'))
data = df[['date','open','close','low','high']]
data=data.values.tolist()
data=split_data(data)
data
candle=Candle(data)
```

```python
candle.kline.render_notebook()
```

```python
import pandas as pd
from sniper.industry.sw import SWManager

sw_manager = SWManager('2010-01-01', SWManager.today)
sw_manager.initialize()
category=sw_manager.category

ticker='801890'
df=sw_manager.read_data(ticker)
df.date=pd.to_datetime(df.date)
df.set_index('date', inplace=True)
df
```

```python
df=sw_manager.read_data(ticker)
data = df[['date','open','close','low','high']]
data.sort_values(by='date',inplace=True)
data=data.values.tolist()
data=split_data(data)
candle=Candle(data)
```
